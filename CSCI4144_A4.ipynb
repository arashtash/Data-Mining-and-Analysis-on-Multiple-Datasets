```python

# Load the dataset
df = pd.read_csv("per-capita-sources-of-protein.csv")


# Rename the columns
df.rename(columns={
    "Vegetal Products | 00002903 || Food available for consumption | 0674pc || grams of protein per day per capita": "Plant_protein_g_per_day",
    "Meat, total | 00002943 || Food available for consumption | 0674pc || grams of protein per day per capita": "Meat_protein_g_per_day",
    "All egg products | 00002744 || Food available for consumption | 0674pc || grams of protein per day per capita": "Egg_protein_g_per_day",
    "Milk | 00002948 || Food available for consumption | 0674pc || grams of protein per day per capita": "Dairy_protein_g_per_day",
    "Fish and seafood | 00002960 || Food available for consumption | 0674pc || grams of protein per day per capita": "Seafood_protein_g_per_day"
}, inplace=True)

#Setting the df to include continent entities only
protein_types = ["Plant_protein_g_per_day", "Meat_protein_g_per_day" , "Egg_protein_g_per_day", "Dairy_protein_g_per_day" , "Seafood_protein_g_per_day"]
continents = ["Africa", "Asia", "Europe", "North America", "Oceania", "South America"]
df = df[df["Entity"].isin(continents)]

# Create a line plot for each protein type
for protein_type in protein_types:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x="Year", y=protein_type, hue="Entity", palette="tab10", linewidth=2)
    plt.title(f"Consumption of {protein_type.split('_')[0]} Protein across continents over the years")
    plt.xlabel("Year")
    plt.ylabel(f"Grams of {protein_type.split('_')[0]} protein consumed per day per capita")
    plt.legend(title="Continent")
    plt.show()

```

```python

# TODO: your code goes here, below. 

df_sustain = pd.read_csv("fish-stocks-within-sustainable-levels.csv")

# Rename columns
df_sustain.rename(columns={
    "14.4.1 Proportion of fish stocks within biologically sustainable levels (not overexploited) (%) | ER_H2O_FWTL || Value | 006121 || percent": "Sustainable_Stocks_Percent",
    "Share of fish stocks that are overexploited": "Overexploited_Stocks_Percent"
}, inplace=True)

# Fill in missing vals
for index, row in df_sustain.iterrows():
    if pd.isna(row["Sustainable_Stocks_Percent"]) and not pd.isna(row["Overexploited_Stocks_Percent"]):
        df_sustain.at[index, "Sustainable_Stocks_Percent"] = 100 - row["Overexploited_Stocks_Percent"]
    elif not pd.isna(row["Sustainable_Stocks_Percent"]) and pd.isna(row["Overexploited_Stocks_Percent"]):
        df_sustain.at[index, "Overexploited_Stocks_Percent"] = 100 - row["Sustainable_Stocks_Percent"]

# Per Capita Source of Protein Preprocessing
df_protein = pd.read_csv("per-capita-sources-of-protein.csv")


# Delete the columns for other protein sources except seafood
df_protein.drop(columns={
    "Vegetal Products | 00002903 || Food available for consumption | 0674pc || grams of protein per day per capita",
    "Meat, total | 00002943 || Food available for consumption | 0674pc || grams of protein per day per capita",
    "All egg products | 00002744 || Food available for consumption | 0674pc || grams of protein per day per capita",
    "Milk | 00002948 || Food available for consumption | 0674pc || grams of protein per day per capita"
}, inplace=True)

#Rename the seafood protein column to something shorter
df_protein.rename(columns={
    "Fish and seafood | 00002960 || Food available for consumption | 0674pc || grams of protein per day per capita": "Seafood_Protein_Consumption_per_capita_per_year"
    }, inplace=True)


df_merged = pd.merge(df_protein, df_sustain, on=["Entity", "Year"], how="inner")

#Drop Cases were sustanable stocks are 0 percent as they seem to be noise
df_merged = df_merged[(df_merged["Sustainable_Stocks_Percent"] != 0)]

#Separating world data and country data
df_world = df_merged[df_merged["Entity"] == "World"]

#Plotting the correlation using scatter chart
plt.figure(figsize=(10, 6))
plt.scatter(df_world["Seafood_Protein_Consumption_per_capita_per_year"], df_world["Sustainable_Stocks_Percent"], color="blue", label="World Data")
plt.title("Seafood Protein Consumption Worldwide vs. Sustainability of Fish Stocks")
plt.xlabel("Seafood Protein Consumption (grams per day per capita)")
plt.ylabel("Proportion of fish stocks within biologically sustainable levels(%)")
plt.legend()
plt.grid(True)

# PLEASE NOTE: I ADDED THIS LINE TO PREVENT MISREPRESENTATION OF THE SCALE. I DID NOT WANT MY GRAPH TO START AT 62%
# ON THE Y-AXIS AS THIS IS A BAD PRACTICE ETHICALLY ACCORDING TO LECTURE 9
plt.ylim(0, 100)
plt.show()


   
```

```python


# TODO: your code goes here, below. Be sure to handle potential errors. 
#   
   
df_environment = pd.read_csv("environmental-factors.csv")

#Renaming columns to shorter names
df_environment.rename(columns={
    "Freshwater use (m3 / kg edible weight)": "Water",
    "Nitrogen (kgN / t edible weight)": "Nitrogen",
    "Land use (m2 / kg edible weight)": "Land",
    "Greenhouse gas emissions (kg / kg edible weight)": "Greenhouse"
    }, inplace=True)

#Adding a separate feature for farmed/wild, extracting whether fish is wild or farmed from Entity or whether it is Chicken
df_environment["Type"] = df_environment["Entity"].apply(lambda x: "Wild" if "wild" in x.lower() else "Farmed")


#Remove the "farmed" and "wild" part from the Entity column for cleaner chart labeling
df_environment["Entity"] = df_environment["Entity"].str.replace(r"\(farmed\)", "", regex=True)
df_environment["Entity"] = df_environment["Entity"].str.replace(r"\(wild\)", "", regex=True)

df_environment_sorted = df_environment.sort_values(by="Greenhouse", ascending=False)

df_chicken = df_environment_sorted[df_environment_sorted["Entity"] == "Chicken"]
df_others = df_environment_sorted[df_environment_sorted["Entity"] != "Chicken"]

plt.figure(figsize=(15, 10))
sns.barplot(data=df_others, x="Entity", y="Greenhouse", hue="Type", palette=["green", "orange"], dodge=0.8)

# Add a horizontal line for the Chicken data
plt.axhline(y=df_chicken["Greenhouse"].values[0], color="blue", linestyle='--', label="Chicken")

plt.title("Greenhouse Gas Emissions by Seafood Species (Blue line is for Chicken)")
plt.xlabel("Seafood Species")
plt.ylabel("Greenhouse Gas Emissions (kg CO2e / kg edible weight)")
plt.xticks(rotation=90)

# Include the Chicken emission line
plt.legend(title="Type")
plt.tight_layout()
plt.show()
```
